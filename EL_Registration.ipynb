```python
########### DOWNSIZE TEMPLATE TO MATCH OUR MAPS ###################################



def resample_img(itk_image, out_size=[128, 128], is_label=False):
    
    # Resample images to 2mm spacing with SimpleITK
    original_spacing = itk_image.GetSpacing() # tamao fsico de un voxel en cada dimensin (mm/pixel)
    original_size = itk_image.GetSize() # # de voxels en cada eje
    out_size = out_size+list(original_size[2:]) # el tamao sera [128,128,z...] z = tamao de dimensin
                                                # z si es que la imagen es de tres dimensiones (se conserva)
    print(original_size)
    print(original_spacing)
    print(out_size)

    out_spacing = [
        original_size[0] * original_spacing[0] / out_size[0],
        original_size[1] * original_spacing[1] / out_size[1],
        # original_size[2] * original_spacing[2] / out_size[2]
    ]

    resample = sitk.ResampleImageFilter()
    resample.SetOutputSpacing(out_spacing)
    resample.SetSize(out_size)
    resample.SetOutputDirection(itk_image.GetDirection())
    resample.SetOutputOrigin(itk_image.GetOrigin())
    resample.SetTransform(sitk.Transform())
    resample.SetDefaultPixelValue(itk_image.GetPixelIDValue())

    if is_label:
        resample.SetInterpolator(sitk.sitkNearestNeighbor)
    else:
        resample.SetInterpolator(sitk.sitkBSpline)

    return resample.Execute(itk_image)
```

```python
############### CORREGISTER T2W IMAGES AND EXTRAPOLATE TO MAPS #################################

# Templates
# plantilla base:
T2w_template_mice_sl3_path = r"/home/enrijuliav/Master/AIBM/CirugiaBariatrica/Template_sl3_corrected.nii.gz"
# plantilla reescalada
SCALED_T2w_template_mice_sl3_path = r"/home/enrijuliav/Master/AIBM/CirugiaBariatrica/Template_sl3_corrected_scaled.nii.gz"
root_dir = r"/home/enrijuliav/Master/AIBM/CirugiaBariatrica/"

template = ants.image_read(T2w_template_mice_sl3_path)
template_scaled = ants.image_read(SCALED_T2w_template_mice_sl3_path)


pattern = os.path.join(root_dir, '**', 'sourcedata', '**', 'anat', '*T2w_masked.nii.gz')
T2w_files = [f for f in glob.glob(pattern, recursive=True) if "corregistered_sl3_final_test" not in f]

# Iterar sobre cada T2w encontrado
for t2w_filename in T2w_files:
    print(f"\n Procesando: {t2w_filename}")

    # Extraer ID del sujeto desde el nombre del archivo 
    try:
        subject_id = [part for part in os.path.basename(t2w_filename).split("_") if part.startswith("sub-")][0]
    except IndexError:
        print(f" No se pudo extraer subject_id de: {t2w_filename}")
        continue

    anat_dir = os.path.dirname(t2w_filename)
    sourcedata_dir = os.path.abspath(os.path.join(anat_dir, "..", ".."))
    resomapper_output_dir = os.path.abspath(os.path.join(sourcedata_dir, ".."))
    derivatives_dir = os.path.join(resomapper_output_dir, "derivatives")

    # Crear carpeta de salida
    corregistered_sl3_dir = os.path.join(resomapper_output_dir, "corregistered_sl3_final_test")
    os.makedirs(corregistered_sl3_dir, exist_ok=True)

    # Leer la imagen T2w y extraer slice 3 
    t2w_img_full = ants.image_read(t2w_filename)

    t2w_slice3 = t2w_img_full[:, :, 2]  

    # Registrar la slice 3 al template
    transformation = ants.registration(fixed=template, moving=t2w_slice3)

    # Guardar imagen T2 registrada
    t2w_out_path = os.path.join(corregistered_sl3_dir, os.path.basename(t2w_filename))
    transformation["warpedmovout"].image_write(t2w_out_path)

    # Mapas a buscar
    map_types = ["MTR", "FA", "AD", "RD", "MD"]

    for map_type in map_types:
        map_pattern = os.path.join(derivatives_dir, "**", f"*{map_type}-processedmap.nii.gz")
        map_files = glob.glob(map_pattern, recursive=True)
        map_files = [f for f in map_files if subject_id in os.path.basename(f)]

        if not map_files:
            print(f"No se encontr mapa {map_type} para: {subject_id}")
            continue

        for map_file in map_files:
            print(f" Registrando mapa {map_type}: {map_file}")
            map_img_full = ants.image_read(map_file)

            map_slice3 = map_img_full[:, :, 2]

            corregistered_map = ants.apply_transforms(
                moving=map_slice3,
                fixed=template_scaled,
                transformlist=transformation["fwdtransforms"]
            )

            output_map_path = os.path.join(corregistered_sl3_dir, os.path.basename(map_file))
            corregistered_map.image_write(output_map_path)





```

# Calculation of Registration Metrics

```python
def calculate_transformation_metrics(fixed_image, warped_image, transform_path):
    """
    fixed_image: ANTsImage (template)
    warped_image: ANTsImage (after registration)
    transform_path: string path to the .nii.gz transform file (displacement field)
    """

    # convert to numpy arrays
    fixed_array = fixed_image.numpy()
    warped_array = warped_image.numpy()

    # cross-correlation 
    # normalize
    fixed_array -= np.mean(fixed_array)
    warped_array -= np.mean(warped_array)

    numerator = np.sum(fixed_array * warped_array)
    denominator = np.sqrt(np.sum(fixed_array ** 2) * np.sum(warped_array ** 2))

    if denominator == 0:
        cross_relation = 0
    else:
        cross_relation = numerator / denominator

    # mean squared error
    mse = np.mean((fixed_array - warped_array) ** 2)

    # jacobian determinant
    try:
        warp_field = ants.image_read(transform_path)
        jacobian_image = ants.create_jacobian_determinant_image(domain_image=fixed_image, tx=warp_field, do_log=False)
        jacobian_array = jacobian_image.numpy()

        jac_min = np.min(jacobian_array)
        jac_max = np.max(jacobian_array)
        jac_mean = np.mean(jacobian_array)
    except Exception as e:
        print(f"Error computing Jacobian: {e}")
        jac_min, jac_max, jac_mean = None, None, None

    return {
        "Cross-correlation": cross_relation,
        "Mean squared error": mse,
        "Jacobian determinant stats": {
            "min": jac_min,
            "max": jac_max,
            "mean": jac_mean
        }
    }
```

```python
def calculate_transformation_metrics(fixed_image, warped_image):
    cc = ants.image_mutual_information(fixed_image, warped_image)
    msd = np.mean((fixed_image.numpy() - warped_image.numpy())**2)
    jacobian = ants.create_jacobian_determinant_image(fixed_image, warped_image, do_log=False)
    jac_stats = {
        'Jacobian Min': np.min(jacobian.numpy()),
        'Jacobian Max': np.max(jacobian.numpy()),
        'Jacobian Mean': np.mean(jacobian.numpy())
    }
    return cc, msd, jac_stats

def visualize_transformation(fixed_image, moving_image, warped_image, registration_result, title):
    try:
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        axes[0].imshow(fixed_image.numpy(), cmap='gray')
        axes[0].set_title('Fixed Image')
        axes[1].imshow(moving_image.numpy(), cmap='gray')
        axes[1].set_title('Moving Image')
        axes[2].imshow(warped_image.numpy(), cmap='gray')
        axes[2].set_title('Warped Image')
        plt.suptitle(title)
        plt.show()

        spacing = 10
        x = np.arange(0, fixed_image.shape[0], spacing)
        y = np.arange(0, fixed_image.shape[1], spacing)
        xv, yv = np.meshgrid(x, y)
        grid_points = np.stack([xv.ravel(), yv.ravel()], axis=1)
        transformed_points = registration_result['fwdtransforms'][0].transform_points(grid_points)

        plt.figure(figsize=(6, 6))
        plt.imshow(fixed_image.numpy(), cmap='gray')
        plt.quiver(grid_points[:, 0], grid_points[:, 1],
                   transformed_points[:, 0] - grid_points[:, 0],
                   transformed_points[:, 1] - grid_points[:, 1],
                   color='red')
        plt.title('Deformation Field')
        plt.show()
    except Exception as e:
        print(f"Error in visualization: {e}")

def compare_registrations(fixed_image, moving1, warped1, reg1, moving2, warped2, reg2):
    cc1, msd1, jac1 = calculate_transformation_metrics(fixed_image, warped1)
    cc2, msd2, jac2 = calculate_transformation_metrics(fixed_image, warped2)

    metrics_df = pd.DataFrame({
        'Metric': ['Cross-Correlation', 'Mean Squared Difference', 'Jacobian Min', 'Jacobian Max', 'Jacobian Mean'],
        'Pre Condition': [cc1, msd1, jac1['Jacobian Min'], jac1['Jacobian Max'], jac1['Jacobian Mean']],
        'Post Condition': [cc2, msd2, jac2['Jacobian Min'], jac2['Jacobian Max'], jac2['Jacobian Mean']]
    })

    print("Comparison of Registration Metrics:")
    print(metrics_df)

    visualize_transformation(fixed_image, moving1, warped1, reg1, "Pre Condition Registration")
    visualize_transformation(fixed_image, moving2, warped2, reg2, "Post Condition Registration")

    print("\nAnalysis:")
    print("""
    The comparison between pre and post condition registrations reveals differences in alignment quality.
    Cross-correlation values indicate how well the intensity patterns match between fixed and warped images.
    Higher values suggest better alignment. Mean squared difference (MSD) quantifies pixel-wise error, with
    lower values indicating more accurate registration.

    Jacobian determinant statistics provide insight into local volume changes. Values near 1 suggest minimal
    distortion, while extreme values may indicate overfitting or anatomical mismatches.

    If post-condition images show higher cross-correlation and lower MSD, they are likely better registered.
    Initial differences in contrast, noise, or anatomical changes between pre and post images can affect
    registration performance. The deformation field visualization helps identify regions of strong warping,
    which may correspond to anatomical shifts or artifacts.

    Overall, cross-correlation and MSD are the most informative metrics for assessing registration quality,
    while Jacobian statistics and deformation fields offer deeper insight into transformation behavior.
    """)
```
